```python
pca_components = [1,2,100,150]
pca_fits=[]

for i in pca_components:
    pca_fits.append(RandomizedPCA(n_components=i,random_state=1999).fit(train.values))
    
#fig = plt.figure(figsize=(8,8))  # figure size in inches
#fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
for p in pca_fits:
    fig = plt.figure(figsize=(8,8))  # figure size in inches
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
    train_pca=p.transform(train.values)
    train_pca_inverse=p.inverse_transform(train_pca)
    for i in range(64):
        ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
        ax.imshow(train_pca_inverse[i].reshape((28,28)), cmap=plt.cm.binary, interpolation='nearest')
        ax.text(0, 7, str(data['label'].values[i]))
```

```python
from sklearn.grid_search import GridSearchCV,RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import train_test_split,cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_pca,data['label'].values, test_size=0.35, random_state=42)
```

```python
rf=RandomForestClassifier(n_estimators=500,random_state=12)
```

```python
scores = cross_val_score(rf,train_pca , data['label'].values , scoring='accuracy', cv=10) 
print('\nAverage Accuracy %.2f +/- %.2f' % (np.mean(scores), np.std(scores)))
```

```python
from sklearn.cross_validation import StratifiedKFold
from sklearn.metrics import accuracy_score
import numpy as np

params = []
scores = []

skfold = StratifiedKFold(y=data['label'].values, n_folds=10, shuffle=False, random_state=1)
for train_idx, test_idx in skfold:
    rf.fit(train_pca[train_idx], data['label'].values[train_idx])
    y_pred = rf.predict(train_pca[test_idx])
    acc = accuracy_score(y_true=data['label'].values[test_idx], y_pred=y_pred)
    scores.append(acc)
```

```python
print('Random forest models:')
for idx, m in enumerate( scores):
    print('%s. Acc: %.2f ' % (idx+1, m))
print('\nAverage Accuracy %.2f +/- %.2f' % (np.mean(scores), np.std(scores)))
```
